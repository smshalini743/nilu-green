```python
#Diurnal variation of Olszankowa,2020

#january

import pandas as pd
df = pd.read_csv("./Olszankowa/2020/Airly-report-2020-01-01T00_00-2020-01-31T00_00.csv")
new = df["From;To;ID 6436 - PM2.5"].str.split(';', 2, expand = True)
df["Date"] = new[0]
df["PM2.5"] = new[2]
PM = []
PM = df["PM2.5"].str.replace('"','')

for i in range(0,len(PM)):
    if PM[i] != '':
        PM[i] = int(PM[i])
    else:
        PM[i] = 0 

Date1 = []
Date1 = df["Date"]
Date1 = list(Date1)

Date1x = []
for j in range(0,len(Date1)):
    abc = Date1[j]
    abcd = abc.replace(abc[5:11], '')          
    Date1x.append(abcd)
Date1x

zippedlist = list(zip(Date1x,PM))
dfobj = pd.DataFrame(zippedlist, columns = ['hourly', 'conc'])
dfobj

dfin = dfobj.groupby(['hourly']).mean()
dfint = dfin.reset_index()
print(dfint)
x = dfint['hourly']
yjan = dfint['conc']
# print(x)
# print(y)

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(18, 12), dpi=150)
plt.plot(x,yjan, linewidth = 3)
plt.xlabel("hour of the day")
plt.ylabel("PM2.5")
plt.title("Sensor Olszankowa - Diurnal Variation(January-2020)")
plt.show()
```

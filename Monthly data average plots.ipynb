```python
#plotting the monthly averages of the sensors 

#Jagiellońska
import pandas as pd
df1 = pd.read_csv("./Jagiellońska/2019/2019-Monthly.csv")
df2 = pd.read_csv("./Jagiellońska/2020/2020-Monthly.csv")
df3 = pd.read_csv("./Jagiellońska/2021/Monthly-jagie.csv")

#print(df1)

new1 = df1["From;To;ID 2851 - PM2.5"].str.split(';', 2, expand = True)
new2 = df2["From;To;ID 2851 - PM2.5"].str.split(';', 2, expand = True)
new3 = df3["From;To;ID 86697 - PM2.5"].str.split(';', 2, expand = True)

df1["Date"] = new1[0]
df1["PM2.5"] = new1[2]
df2["Date"] = new2[0]
df2["PM2.5"] = new2[2]
df3["Date"] = new3[0]
df3["PM2.5"] = new3[2]

PM19 = []
PM20 = []
PM21 = []

PM19 = df1["PM2.5"].str.replace('"', '')
PM20 = df2["PM2.5"].str.replace('"', '')
PM21 = df3["PM2.5"].str.replace('"', '')

for i in range(0,len(PM19)):
    if PM19[i] != '':
        PM19[i] = int(PM19[i])
    else:
        PM19[i] = 0
for i in range(0,len(PM20)):
    if PM20[i] != '':
        PM20[i] = int(PM20[i])
    else:
        PM20[i] = 0
for i in range(0,len(PM21)):
    if PM21[i] != '':
        PM21[i] = int(PM21[i])
    else:
        PM21[i] = 0
        
Date1 = []
Date1 = df1["Date"]
Date1x = []
for j in range(0,len(Date1)):
    abc = Date1[j].replace("/2019", "")
    Date1x.append(abc) 

# print(Date1x)
# print(PM19)
# print(PM20)
# print(PM21)
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(12, 8), dpi=100)

plt.plot(Date1x,PM19, color ='r', label ='2019')
plt.plot(Date1x,PM20, color = 'b', label = '2020')
plt.plot(Date1x,PM21, color = 'y', label = '2021')
plt.xticks()
plt.yticks()
plt.xlabel('Months')
plt.ylabel('PM2.5 conc')
plt.title('Sensor Jagiellońska')

plt.legend()
plt.show()
```

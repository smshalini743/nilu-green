```python
#daily variation of PM10 data in Jagie

import pandas as pd
dailydf = pd.read_csv("../Jagie-Jan-Oct(PM10) new.csv")

#splitting the columns
new = dailydf["PM10"].str.split(';', 2, expand = True)
#print(new)

dailydf["Date"] = new[0]
dailydf["PM10 conc"] = new[2]
#print(dailydf)

newconc = dailydf["PM10 conc"].str.replace('"', '')
#print(newconc)

newdate = dailydf["Date"]
#print(newdate)

#print(dailydf.iloc[0,1])
```

```python
x = []
y = []

x = newdate.tolist()
y = newconc.tolist()

#for removing the /2021 from date for plotting clearly
z = []
for j in range(0,len(x)):
    abc = x[j].replace("/2021", "")
    z.append(abc)

#print(z)

#converting string to int and the NaN values to zero
for i in range(0, len(y)):
    if y[i] != '':
        y[i] = int(float(y[i]))
    else:
        y[i] = 0
        
#the two lists of date and conc were added to a single dataframe         
# zippedlist = list(zip(z,y))
# dfobj = pd.DataFrame(zippedlist, columns = ['dailydata', 'conc'])
# dfobj

#running a loop for making a new list for a single month
month1 = []
month2 = []
month3 = []
month4 = []
month5 = []
month6 = []
month7 = []
month8 = []
month9 = []
month10 = []

for m in z:
    if m[3:5] == "01":
        month1.append(m)
    elif m[3:5] == "02":
        month2.append(m)
    elif m[3:5] == "03":
        month3.append(m)
    elif m[3:5] == "04":
        month4.append(m)
    elif m[3:5] == "05":
        month5.append(m)
    elif m[3:5] == "06":
        month6.append(m)
    elif m[3:5] == "07":
        month7.append(m)
    elif m[3:5] == "08":
        month8.append(m)
    elif m[3:5] == "09":
        month9.append(m)
    elif m[3:5] == "10":
        month10.append(m)
    
# print(month1)
# print(month2)
# print(month3)
# print(month4)
# print(month5)
# print(month6)
# print(month7)
# print(month8)
# print(month9)
# print(month10)


janconc = []
febconc = []
marconc = []
aprconc = []
mayconc = []
junconc = []
julconc = []
augconc = []
sepconc = []
octconc = []

#running a loop for getting the conc value of each month
for n in range(0, len(y)):
    if n < 31:
        janconc.append(y[n])
    elif n < 59:
        febconc.append(y[n])
    elif n < 90:
        marconc.append(y[n])
    elif n < 120:
        aprconc.append(y[n])
    elif n < 151:
        mayconc.append(y[n])
    elif n < 181:
        junconc.append(y[n])
    elif n < 212:
        julconc.append(y[n])
    elif n < 243:
        augconc.append(y[n])
    elif n < 273:
        sepconc.append(y[n])
    elif n < 304:
        octconc.append(y[n])
        

# print(janconc)
# print(febconc)
# print(marconc)
# print(aprconc)
# print(mayconc)
# print(junconc)
# print(julconc)
# print(augconc)
# print(sepconc)
# print(octconc)
   

```

```python
#plotting the daily data for a month

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(18, 12), dpi=150)

plt.plot(month1, janconc)
plt.xlabel("Days")
plt.ylabel("PM10 conc")
plt.title("Daily variation of PM10 for January - 2021")
plt.show()
```

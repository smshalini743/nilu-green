```python
#code for Poland Sensors (Temporal varitaion)

import pandas as pd

#sta1 stands for station 1 - Jagie
sta1 = pd.read_csv("./Data/Jagie-Jan-Oct(PM10)-Monthly new.csv")

df = sta1

print(df.head(2))
```

```python
#the first object in the column
#firstobject = sta1.iloc[0]
#display(firstobject)

#converted the dataframe into a string #though not required here
#dftos = df["PM10"].to_string()
#print(type(dftos))

#dftos = dftos.replace('\"', ' ')
#print(dftos)


#splitting the columns 
new = df["PM10"].str.split(';', 2, expand = True)
#print(new)


#making a separate column for the first date
df["Date"] = new[0] 

#making separate column for the pm10 conc
df["PM10 conc"] = new[2]

#the new columns are added to the dataframe 
df

#removal of apostrophe from the concentration
newconc = df["PM10 conc"].str.replace('"', ' ')
print(newconc)

newdate = df["Date"]
print(newdate)



```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(12, 10), dpi=100)
#made two lists to add the date and conc columns 
x = []
y = []

x = newdate.tolist()
y = newconc.tolist()

print(y)

#converting string data to integer
for i in range(0, len(y)):
    y[i] = int(y[i])

#print(x)
#print(y)

plt.plot(x,y)
plt.xlabel("Months")
plt.ylabel("PM10 conc")
plt.title("Monthly variation of PM10")
plt.show()


```

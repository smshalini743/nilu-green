```python
#diurnal variation of Polskiej Organizacji Wojskowej - 2020

#october
import pandas as pd
dfoct = pd.read_csv("./Polskiej Organizacji Wojskowej/2020/Airly-report-2020-10-01T00_00-2020-10-31T00_00.csv")
newoct = dfoct["From;To;ID 31103 - PM2.5"].str.split(';', 2, expand = True)
dfoct["Date"] = newoct[0]
dfoct["PM2.5"] = newoct[2]
PMoct = []
PMoct = dfoct["PM2.5"].str.replace('"','')

for i in range(0,len(PMoct)):
    if PMoct[i] != '':
        PMoct[i] = int(PMoct[i])
    else:
        PMoct[i] = 0

Date10 = []
Date10 = dfoct["Date"]
Date10 = list(Date10)

Date10x = []
for j in range(0,len(Date10)):
    abc = Date10[j]
    abcd = abc.replace(abc[5:11], '')          
    Date10x.append(abcd)

zippedlist = list(zip(Date10x,PMoct))
dfoctober = pd.DataFrame(zippedlist, columns = ['hourly', 'conc'])
dfoctober

dfoctober1 = dfoctober.groupby(['hourly']).mean()
dfoctober2 = dfoctober1.reset_index()
print(dfoctober2)

x = dfoctober2['hourly']
yoct = dfoctober2['conc']
# print(x)
# print(y)

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(18, 12), dpi=150)
plt.plot(x,yoct, linewidth = 3)
plt.xlabel("hour of the day")
plt.ylabel("PM2.5")
plt.title("Sensor Polskiej Organizacji Wojskowej - Diurnal Variation(October-2020)")
plt.show()
```
